## E) Pairwise Scatter Plots of Integer Variables in Supply chain dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

# Set style for the plots
sns.set_theme(style="whitegrid")

# Create figure for scatter plots
fig = plt.figure(figsize=(15, 12))

# Filter integer columns
int_columns = [col for col in df.columns if df[col].dtypes in ["int64", "float64"]]

# Define the maximum number of plots and select column pairs
max_plots = 12  # Adjust based on desired number of plots
combinations = list(itertools.combinations(int_columns, 2))[:max_plots]

# Generate scatter plots for each selected pair of columns
for i, (col1, col2) in enumerate(combinations):
    plt.subplot(4, 3, i + 1)  # Arrange subplots in a grid
    sns.scatterplot(data=df, x=col1, y=col2, alpha=0.7, color="blue")  # Add transparency and color
    plt.title(f'{col1} vs {col2}', fontsize=10)  # Title for each scatter plot
    plt.xlabel(col1, fontsize=8)  # Label x-axis
    plt.ylabel(col2, fontsize=8)  # Label y-axis
    plt.tick_params(axis='both', which='major', labelsize=7)  # Adjust tick label size

# Adjust layout for better spacing and display the plot
fig.tight_layout(pad=2.0)
plt.show()
```

## 20. Pareto chart

-  A Pareto chart helps identify the top contributors to defects by focusing on the 80/20 rule, which means that 80% of the defects often come from 20% of the causes. 
-  To create a Pareto chart, we need to calculate the frequency of defects for each category (e.g., Supplier name, Product type) and then visualize it in a descending order with both a bar chart and a cumulative line plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Grouping data by 'Supplier name' and summing defect rates (or frequency, depending on the context)
defects_by_supplier = df.groupby('Supplier name')['Defect rates'].sum().reset_index()

# Sorting by Defect rates in descending order
defects_by_supplier = defects_by_supplier.sort_values(by='Defect rates', ascending=False)

# Calculating cumulative percentage
defects_by_supplier['Cumulative Defects'] = defects_by_supplier['Defect rates'].cumsum()
defects_by_supplier['Cumulative Percentage'] = 100 * defects_by_supplier['Cumulative Defects'] / defects_by_supplier['Defect rates'].sum()

# Creating the Pareto Chart
fig, ax1 = plt.subplots(figsize=(6,4))

# Bar plot for Defect rates
sns.barplot(
    x='Supplier name', 
    y='Defect rates', 
    data=defects_by_supplier, 
    color='skyblue', 
    ax=ax1
)

# Adding the Cumulative Percentage line
ax2 = ax1.twinx()
ax2.plot(
    defects_by_supplier['Supplier name'], 
    defects_by_supplier['Cumulative Percentage'], 
    color='red', 
    marker='o', 
    linestyle='-', linewidth=2, markersize=6
)

# Titles and labels
ax1.set_title('Pareto Chart for Defects by Supplier', fontsize=14)
ax1.set_xlabel('Supplier Name', fontsize=12)
ax1.set_ylabel('Defect Rates', fontsize=12)
ax2.set_ylabel('Cumulative Percentage (%)', fontsize=12)


plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
